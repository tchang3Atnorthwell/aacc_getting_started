Challenge #3

Determine the pairwise correlation of all analytes;
Plot the pairwise correlation (i.e., regression scatterplot) of all analytes; and

Determine which analytes are most highly correlated.

```python
# seaborn has tools to quickly look at all of the data

sns.pairplot(chem8)
```

```python
# sns.pairplot plots on 2 axes

sns.pairplot(chem8[['Na','K']], kind='reg')
```

```python
# but we'll downsample to make the correlations more clear and color the regression line

sns.pairplot(chem8[['Na','K']].sample(n=100), kind='reg', plot_kws={'line_kws':{'color':'red'}})
```

```python
# now we can do all of the analytes

sns.pairplot(chem8[chem8.columns[2:]].sample(n=100), kind='reg', plot_kws={'line_kws':{'color':'red'}})
```

Challenge #4: 

Calculate the anion gap for each row; and
Find the subset of the data for which the anion gap is "abnormal," which we can define as those values greater than 16 mmol/L.

```python
# calculations are straightforward in pandas

chem8['AnionGap']=chem8['Na']+chem8['K']-chem8['Cl']-chem8['CO2']
chem8
```

```python
# you can add limits to select subsets of data

chem8[chem8['AnionGap']>16]
```
